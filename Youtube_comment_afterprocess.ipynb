```python
import re

# Compile regex patterns
emoji_pattern = re.compile(
    "["
    u"\U0001F600-\U0001F64F"  # emoticons
    u"\U0001F300-\U0001F5FF"  # symbols & pictographs
    u"\U0001F680-\U0001F6FF"  # transport & map symbols
    u"\U0001F1E0-\U0001F1FF"  # flags
    u"\U00002700-\U000027BF"  # dingbats
    u"\U0001F900-\U0001F9FF"  # supplemental symbols
    u"\U00002600-\U000026FF"  # miscellaneous symbols
    "]+", flags=re.UNICODE
)

url_pattern = re.compile(r"http\S+|www\S+")
email_pattern = re.compile(r"\S+@\S+")
phone_pattern = re.compile(r"\+?\d[\d\s\-]{7,}\d")
currency_pattern = re.compile(r"[‡ß≥$‚Ç¨¬£¬•‚Çπ]")

# Cleaning function
def clean_comment(text):
    text = str(text)
    text = emoji_pattern.sub('', text)          # Remove emojis
    text = url_pattern.sub('', text)            # Remove URLs
    text = email_pattern.sub('', text)          # Remove emails
    text = phone_pattern.sub('', text)          # Remove phone numbers
    text = currency_pattern.sub('', text)       # Remove currency symbols
    text = re.sub(r'\s+', ' ', text).strip()    # Remove extra whitespace
    return text

# Apply to the 'comment' column
no_dupe_comments.loc[:, 'Comment'] = no_dupe_comments['Comment'].apply(clean_comment)
```

```python
# Step 1: Convert all values to string, remove 'Z' if present
no_dupe_comments.loc[:, 'Timestamp'] = no_dupe_comments['Timestamp'].astype(str).str.replace('Z', '', regex=False)

# Step 2: Convert the cleaned string to datetime
no_dupe_comments.loc[:, 'Timestamp'] = pd.to_datetime(no_dupe_comments['Timestamp'], errors='coerce')

# Step 3: Extract date and time
no_dupe_comments.loc[:, 'Date'] = no_dupe_comments['Timestamp'].dt.date
no_dupe_comments.loc[:, 'Time'] = no_dupe_comments['Timestamp'].dt.time
```
